```python
import os
import pandas as pd
from datetime import datetime
import sys as sys

excel_file_path = './files/PrismaSeperation_Aamazon/Amazon_Automated_Vendor_Placements.xlsx'

    
        
def process_excel_file(file_path):
    # Read the Excel file
    try:
        df = pd.read_excel(file_path)
    except FileNotFoundError:
        print(f"Error: The file {file_path} does not exist.")
        return
            
        

    # Get unique values from the first column
    unique_values = df.iloc[:, 0].unique()

    # Get the current year
    current_year = datetime.now().year

    # Define the directory to save files
    save_directory = "./files/PrismaSeperation_Aamazon/Results"

    camp_name = input(prompt = "What is the Campaign Name?")

    # Process each unique value
    for value in unique_values:
        # Create a DataFrame for the current unique value
        subset_df = df[df.iloc[:, 0] == value]

        # Print or process the subset DataFrame as needed
        print(f"Processing unique value: {value}")
        print(subset_df)

        # Save the subset DataFrame to a separate Excel file with current year + unique value
        file_name = f"{current_year}_{value}_{camp_name}.xlsx"
        file_path = os.path.join(save_directory, file_name)
        subset_df.to_excel(file_path, index=False)
        print(f"Saved to file: {file_path}")


# Call the function to process the Excel file
process_excel_file(excel_file_path)
```
